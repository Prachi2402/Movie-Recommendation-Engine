# Data Preprocessing 

```python
#convert into lowercase & remove punctuations, numbers etc
df['clean_plot'] = df['Plot'].str.lower()
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('[^a-zA-Z]',' ',x)) #all the other characters except alphabets will be replaced by space.
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('\s+',' ',x)) #replaced extra spaces with a single space 
df['clean_plot']
```

```python
import nltk
nltk.download('punkt')
```

```python
# tokenize the sentence 
df['clean_plot'] = df['clean_plot'].apply(lambda x: nltk.word_tokenize(x))
df['clean_plot']
```

```python
import nltk
nltk.download('stopwords')
```

```python
# remove stop words 
stop_words = nltk.corpus.stopwords.words('english')
plot = []
for sentence in df['clean_plot']:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) > 3:
            temp.append(word)
    plot.append(temp)
plot
```

```python
df['clean_plot'] = plot
df['clean_plot']
```

```python
df['Genre'] = df['Genre'].apply(lambda x: x.split(','))
df['Actors'] = df['Actors'].apply(lambda x: x.split(',')[:4])
df['Director'] = df['Director'].apply(lambda x: x.split(','))
```

```python
def clean(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ',''))
    return temp
```

```python
df['Genre'] = [clean(x) for x in df['Genre']]
df['Actors'] = [clean(x) for x in df['Actors']]
df['Director'] = [clean(x) for x in df['Director']]
```

```python
df['Actors'][0]
```

```python
#combining all the columns data
columns = ['clean_plot', 'Genre', 'Actors', 'Director']
l = []
for i in range(len(df)):
    words = ''
    for col in columns: 
        words += ''.join(df[col][i]) + ' '
    l.append(words)
l
```

```python
df['clean_input'] = l
df = df[['Title', 'clean_input']]
df.head()
```

# Feature Extraction 

```python
pip install sklearn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df['clean_input'])
```

```python
#create cosine similarity matrix 
from sklearn.metrics.pairwise import cosine_similarity 
cosine_sim = cosine_similarity(features, features)
print(cosine_sim)
```

# Movie Recommendation

```python
index = pd.Series(df['Title'])
index.head()
```

```python
def recommend_movies(title):
    movies = []
    idx = index[index == title].index[0]
    #print(idx)
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)
    #print(top10)
    
    for i in top10:
        movies.append(df['Title'][i])
    return movies
```

```python
#recommending similar movies based on the plot, directors, actors
recommend_movies('The Shawshank Redemption')
```

```python
recommend_movies('The Avengers')
```
